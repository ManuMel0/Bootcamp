```python
funcionarios = [] # todos os funcionarios serão salvos nessa lista
id = 0 # esse id é o que será o id de cada usuário onde tera o incremento toda vez que a funçao addFuncionario for chamada



def addFuncionario(): # função que é responsavel pela adição de um novo funcionario

  global id # chamando a variável global Id para assim poder fazer o incremento e cada funcionário ter um id unico

  while True: # esse while juntamente com o try é responsavel para que se algum dado digitado pelo usuario estiver incorreto, ele repete todo o procedimento de coleta de dados do funcionário
     try:
       # coletando dados para o cadastro:
       print('\nDigite a seguir os dados referente ao usuario que será cadastrado:')
       print(f'ID: {id}')
       nome = input('Nome Completo: ').title()
       idade = int(input('Idade: '))
       cidade = input('Cidade: ').title()
       estado = input('Estado: ').title()
       escolaridade = input('Escolaridade: ').title()
       email = input('E-mail: ')
       cargo = input('Cargo: ').title()
       salario = float(input('Salário: '))

       funcionarios.append([id, nome, idade, cidade, estado, escolaridade, email, cargo, salario]) # adicionando uma lista dentro de funcionarios com os dados do funcionário digitado
       id +=1 # fazendo o incremento para cada funcionario ter um ID uníco
       break

     except: # caso alguma infomação digitada não seja a esperada, vai rodar o except onde ficara em loop até digitar 1 para tentar novamente ou 2 para voltar ao menu.
       i = int(input('Algum dado está incorreto, digite 1 para tentar novamente ou 2 para voltar para o menu principal: '))

       # redirecionando o usuário quando digitar 1 ou 2
       if i == 1: # reinicia o cadastro do funcionário
        addFuncionario()
        break
       else:
        menu() # volta para o menu
        break




def listarFuncionario(idFuncionarioSelecionado): # essa função lista somente um funcionário e também retorna os dados desse funcionario e seu indice

    funcionarioSelecionado = [] # lista com os dados do funcionario selecionado
    idFuncionarioSelecionado = int(idFuncionarioSelecionado) # variável com o id do funcionário

    for i,v in enumerate(funcionarios): # percorrendo a lista funcionarios para encontrar qual funcionário corresponde ao ID digitado
      if v[0] == idFuncionarioSelecionado: # v[0] corresponde ao ID do funcionario
        funcionarioSelecionado = v # variável atualizada com os dados do funcionário correspondente ao ID digitado
        indiceFuncionarioSelecionado = i # variável atualizada com o indice do funcionário correspondente ao ID digitado

    if(funcionarioSelecionado == []): # se o ID digitado não corresponder a nenhum funcionário, esse if será acionado e retorna para o menu
      print('Funcionario não encontrado, Digite 5 no menu para listar todos os funcionários e verificar se esse funcionário  se encontra na lista\n')
      menu()
    else: # esse else será executado se a variavel com os dados não esteja vazia
      print('\nDados dos funcionários Selecionado:')    # aqui imprimi todos os dados do funcionário corresponde ao ID digitado
      print('Funcionário ID:', funcionarioSelecionado[0])
      print('1- Nome Completo:', funcionarioSelecionado[1])
      print('2- Idade:', funcionarioSelecionado[2])
      print('3- Cidade:', funcionarioSelecionado[3])
      print('4- Estado:', funcionarioSelecionado[4])
      print('5- Escolaridade:', funcionarioSelecionado[5])
      print('6- E-mail:', funcionarioSelecionado[6])
      print('7- Cargo:', funcionarioSelecionado[7])
      print('8- Salário:', funcionarioSelecionado[8])
      print()
      return funcionarioSelecionado, indiceFuncionarioSelecionado   # esse return é importante para a funcao de remover e de atualizar funcionario
                                                                    #   pois utilizando ele, não sera necessário repetir todo esse codigo acima
                                                                    #                    nas outras duas funções.



def listarFuncionarios(funcionarios):  # função responsavel por listar todos os funcionários
    print('\nDados dos funcionários cadastrados:')
    for funcionario in funcionarios: # loop para percorrer cada funcionário e imprimir ele na tela
        print('Funcionário ID:', funcionario[0])
        print('Nome Completo:', funcionario[1])
        print('Idade:', funcionario[2])
        print('Cidade:', funcionario[3])
        print('Estado:', funcionario[4])
        print('Escolaridade:', funcionario[5])
        print('E-mail:', funcionario[6])
        print('Cargo:', funcionario[7])
        print('Salário:', funcionario[8])
        print()




def removerFuncionario(idFuncionarioSelecionado): # função responsavel por remover um funcionário

  try:
    funcionarioSelecionado, indiceFuncionarioSelecionado = listarFuncionario(idFuncionarioSelecionado)
  except:
    return
  '''
  Utilizando a função de listar apenas um funcionário, conseguimos imprimir os dados desse funcionário para assim o usuário confirmar se
  realmente é esse o funcionário desejado e ainda conseguimos guarda o indice desse funcionário na variável indiceFuncionarioSelecionado e
  os dados desse funcionário em funcionarioSelecionado. O importante para fazer a remoção dele será somente a variável com o indice do funconario.

  Quando o usuario digita um ID que não corresponde a nenhum funcionário, o return de listarFuncionario(idFuncionarioSelecionado) dará erro,
  por isso tem esse try. Já foi feito o tratamente desse erro na funcao de imprimir um funcionário porém quando o programa encerra ele ainda da erro
  sendo assim necessário esse try aqui.
  '''

  while True: # esse while evita que o usuario quebre o código digitando um valor que não é esperado
    removerFuncionario = int(input('Realmente deseja exluir esse funcionário (digite 1 para sim e 2 para não): ')) # confirma se o usuario deseja excluir o funcionário
    if removerFuncionario == 1: # remove o funcionário
      funcionarios.pop(indiceFuncionarioSelecionado)
      break
    elif removerFuncionario == 2: # volta para o menu e não exclui o funcionário
      menu()
      break
    else: # se o valor não for 1 ou 2, será pedido novamente que o usuario digite uma dessas duas variáveis
      print("Erro: Entrada inválida. Por favor, digite um número válido.\n")




def atualizarFuncionario(idFuncionarioSelecionado):  # função que atualiza o funcionário selecionado

  try:
    funcionarioSelecionado, indiceFuncionarioSelecionado = listarFuncionario(idFuncionarioSelecionado) # imprimi o funcionário selecionado e retorna os dados dele
                                                                                                       # na variável funcionarioSelecionado e seu indice na variável
                                                                                                       # indiceFuncionarioSelecionado
  except:
    return
  '''
  Quando o usuario digita um ID que não corresponde a nenhum funcionário, o return de listarFuncionario(idFuncionarioSelecionado) dará erro,
  por isso tem esse try. Já foi feito o tratamente desse erro na funcao de imprimir um funcionário porém quando o programa encerra ele ainda da erro
  sendo assim necessário esse try aqui.
  '''


  # essa variável indiceDeAlterar guarda o indice que corresponde ao que o usuario deseja alterar nos dados do funcionário
  indiceDeAlterar = int(input('Digite o número respectivo ao que vc deseja alterar (1 para alterar nome, 2 para idade e assim por diante): '))

  while True: # esse while evita que o usuario quebre o código digitando um valor que não é esperado

    # novoValor guarda o valor novo digitado nessa variável. Se o usuário apertar apenas apertar Enter, será salvo o valor anterior
    novoValor = input(f'Digite o novo valor (ou pressione Enter para manter "{funcionarioSelecionado[indiceDeAlterar]}"): ') or funcionarioSelecionado[indiceDeAlterar]

    try: # try utilizado pois a variável idade é int e o salário é float. Nesse try se o valor não for digitado correto e acabar dando erro na transformação de alguma dessas variáveis,
         #       será acionado o except fazendo com que o usuário digite um novo valor até parar de dar erro nessa transformação de str para int ou float

      if indiceDeAlterar == 2: # se foi mudado a idade, transforma o dado digitado em int para guardar em idade no type correto dessa variável
        funcionarios[indiceFuncionarioSelecionado][indiceDeAlterar] = int(novoValor)
        break

      elif indiceDeAlterar == 8: # se foi mudado o salário, transforma o dado digitado em float para guardar em salário no type correto dessa variável
        funcionarios[indiceFuncionarioSelecionado][indiceDeAlterar] = float(novoValor)
        break

      else: # se o novo valor não for idade ou salário, os dados do fucionarios são atualizado sem nenhuma transformação pois os dados restantes são todos strings
        funcionarios[indiceFuncionarioSelecionado][indiceDeAlterar] = novoValor
        break

    except:
        print("Erro: Entrada inválida. Por favor, digite um número válido.\n")




def menu(): # função com o menu de todas as opções possiveis do programa
   print("\n===== MENU =====")
   print("1. Adicionar Funcionário;")
   print("2. Atualizar dados de um funcionário já cadastrado;")
   print("3. Visualizar cadastro de um funcionário específico;")
   print("4. Remover cadastro de funcionário;")
   print("5. Listar Funcionários;")
   print("6. Sair.")

   # op == opção, ou seja, recebe a opção selecionada pelo usuario e em seguida executa a função correspondente a opção selecionada
   op = int(input("Digite o número referente a uma das opções citadas: "))
   if op == 1:
     addFuncionario()

   elif op == 2:
     idFuncionarioSelecionado = input('Digite o ID do funcionario que deseja alterar os dados: ')
     atualizarFuncionario(idFuncionarioSelecionado)

   elif op == 3:
     idFuncionarioSelecionado = input('Digite o ID do funcionario que deseja visualizar: ')
     listarFuncionario(idFuncionarioSelecionado)

   elif op == 4:
     idFuncionarioSelecionado = input('Digite o ID do funcionario que deseja excluir: ')
     removerFuncionario(idFuncionarioSelecionado)

   elif op == 5:
     listarFuncionarios(funcionarios)

   elif op == 6:
     print("Saindo do sistema...")
     return

   else:
     print("Opção inválida. Tente novamente.")
     menu()
```

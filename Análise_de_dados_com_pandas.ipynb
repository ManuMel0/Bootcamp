```python
import pandas as pd

df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQXJcDO6iN8cJTdXtvX7UHC41NwyY7w0K_4Y7mVY79Ai3WVDCr7ZdRLcVxdOPpz8A/pub?gid=1469769408&single=true&output=csv')
display(df)
```

```python
# Convertendo a coluna "Demissão" para valores numéricos (0 para não e 1 para sim)
df['Demissão'] = df['Demissão'].map({'Sim': 1, 'Não': 0})

display(df)
```

```python
# Selecionando apenas as colunas numéricas
dados_numericos = df.select_dtypes(include='int64')

# Calculando a correlação entre "demissões" e todas as outras colunas numéricas
correlacoes = dados_numericos.corrwith(df['Demissão'])

# Exibindo as correlações
print(correlacoes)
```

```python
# calculando a porcentagem de relação
porcentagem_correlacao = round((correlacoes ** 2) * 100,2)
print(porcentagem_correlacao.sort_values(ascending=False))
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQXJcDO6iN8cJTdXtvX7UHC41NwyY7w0K_4Y7mVY79Ai3WVDCr7ZdRLcVxdOPpz8A/pub?gid=1469769408&single=true&output=csv')

colunas_object = df.select_dtypes(include=['object'])
colunas_object.info()
```

# Filtrado por pessoas que pediram demissão

```python
demissaoSim = colunas_object.loc[df['Demissão'] == 'Sim']
porcentagemGroupby = demissaoSim.groupby(['Viagem de negócios'])[['Viagem de negócios']].count()/demissaoSim['Viagem de negócios'].count()*100
round(porcentagemGroupby, 2)
```

```python
demissaoSim = colunas_object.loc[df['Demissão'] == 'Sim']
porcentagemGroupby = demissaoSim.groupby(['Departamento'])[['Departamento']].count()/demissaoSim['Departamento'].count()*100
round(porcentagemGroupby, 2)
```
